```python
# Creating sample DataFrame with numeric columns

data = {'Numeric_column1':[5,15,8],
        'Numeric_column2':[10,50,80],
        'Numeric_column3':[100,200,400]}
df = pd.DataFrame(data)


mean_value = df.mean()
print(mean_value)

std_value = df.std()
print("std: \n",std_value)

mode_value = df.mode()
print("mode: \n",mode_value)
```

```python
# Assuming 'df' is your DataFrame with a 'Date' Column

data = {'Date':['2023-01-01','2023-02-15','2023-03-20']}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Extracting weekday and weekend information.

df = pd.DataFrame({'Date':pd.date_range(start='2023-01-01', periods = 5)})
df['Weekday'] = df['Date'].dt.weekday

# df['Date'].dt.weekday access the date column as a datetime object(dt accessor) and
# retrevies the weekday ie. (monday = 0, sunday = 6) for each date.

df['IsWeekend'] = df['Date'].dt.weekday // 5 == 1

# df['Date'].dt.weekday //5 calculates the integer division of the weekday index by 5.
# This operation results in for surday(5 divide by 5) and 0 for weekdays

print(df[['Date','Weekday','IsWeekend']])
```

```python
# Shifting dates forward or backward

df['Date'] = pd.to_datetime(df['Date'])
df['PreviousDate'] = df['Date'] - pd.Timedelta(days = 1)
df["NextDate"] = df['Date'] + pd.Timedelta(days = 1)

print(df[['Date','PreviousDate','NextDate']])
```
