# Pandas joining Technique.
Pandas provides various joining techniques, such as merging, joining, and
concatenating, which allow datasets to be combined using one or more keys. Each
method has unique behaviors and applications.

<!-- Concatenate:  -->
It appends DataFrames vertically or horizontally, offering a
straightforward way to combine datasets with distinct columns or indices without
regard for overlapping keys or index values.


<!-- Merge: -->
It combines DataFrames by aligning columns with shared keys, allowing
for detailed control over overlapping column names and the use of multiple keys.

<!-- Join: -->
It aligns DataFrames based on their index values, making it ideal for
coordinating data with corresponding indices.
These techniques are essential for integrating and analyzing different datasets. They
enable a thorough understanding

```python
# Pandas Concatenate

df1 = pd.DataFrame(
    {
        "A":["A0","A1","A2","A3"],
        "B":["B0","B1","B2","B3"],
        "C":["C0","C1","C2","C3"],
        "D":["D0","D1","D2","D3"],
    },
    index=[0,1,2,3],
)


df2 = pd.DataFrame(
    {
        "A":["A4","A5","A6","A7"],
        "B":["B4","B5","B6","B7"],
        "C":["C4","C5","C6","C7"],
        "D":["D4","D5","D6","D7"],
    },
    index=[0,1,2,3],
)

df3 = pd.DataFrame(
    {
        "A":["A8","A9","A10","A11"],
        "B":["B8","B9","B10","B11"],
        "C":["C8","C9","C10","C11"],
        "D":["D8","D9","D10","D11"],
    },
    index=[0,1,2,3],
)

frames = [df1,df2,df3]
Result = pd.concat([df1,df2,df3])
Result
```

```python
# To concatinate by rows(Vertical)

Result_rows = pd.concat([df1,df2,df3], axis=0)
Result_rows
```

```python
# To Concaticate by Columns(Horizontal)

Result_columns = pd.concat([df1,df2,df3], axis=1)
Result_columns
```

# Pandas Merge DataFrames (It is for columns wise merging, it needs common columns)
Utilize the pd.merge() method to merge DataFrames based on specific keys or
columns.
Specify the join type in Pandas merge, which controls how rows from two
DataFrames are combined.
This ensures data alignment and prevents unintended outcomes.
Choose the appropriate how parameter to specify the type of join.
Specify the on parameter to indicate the column(s) to merge on.
Types of Pandas Join
There are various join logics available to merge Pandas DataFrames:

<!-- Full Outer Join:  -->
It merges all rows from both DataFrames, using NaN to fill in
missing values when no match is found.
<!-- All the values from both dataframe -->

<!-- Inner Join:  -->
It combines matching rows from DataFrame 1 and DataFrame 2 based
on a common key column.
<!--Only the common values from both the dataframe  -->

<!-- Right Join: -->
It retains all rows from the right DataFrame, merges on common keys,
and fills missing values with NaN.
<!--All the values from right dataframe and the common value of 2 dataframe  -->

<!-- Left Join:  -->
It retains all rows from the left DataFrame, merging matching rows from
the right and filling unmatched values with NaN.
<!-- All the value of the left dataframe and the common value of 2 dataframe -->

<!-- Cross:  -->
It creates the cartesian product of the rows of both frames.

```python
left = pd.DataFrame(
    {
    "key1": ["K0", "K0", "K1", "K2"],
    "key2": ["K0", "K1", "K0", "K1"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
    }
)
right = pd.DataFrame(
    {
    "key1": ["K0", "K1", "K1", "K2"],
    "key2": ["K0", "K0", "K0", "K0"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
    }
)

print(left)
```

```python
print(right)
```

```python
Result_outer = pd.merge(left, right, how = 'outer', on =['key1','key2'])
Result_outer
```

```python
Result_inner = pd.merge(left,right[['key1','key2','C']],how ='inner',left_on =['key1','key2'], right_on =['key1','key2'])
Result_inner

# Only C is mapped from Right DataFrame.
# D is not mapped
```

```python
Result_left = pd.merge(left, right, how = 'left', on=['key1','key2'])
Result_left
```

```python
Result_right = pd.merge(left, right, how = 'right', on=['key1','key2'])
Result_right
```

```python
Result_cross = pd.merge(left,right , how = 'cross')
Result_cross
```

# Aggregating Data (Summarizing)
Aggregating data involves summarizing or grouping data based on specific criteria.
This is useful for creating meaningful insights and reducing data dimensionality.
Common aggregation functions include average(mean), median, minimum(min),
maximum(max), sum, standard deviation(std), variance(var), and count

```python
# Creating a DataFrame with a 'Category' Column and a 'Value' Column

txs = pd.DataFrame({'cust' :['A','B','A','B','A'],
      'txn_amount' : [100,200,300,400,500]
      })
txs
```

```python
# We need to aggreate the total txn_amount of A and B

# Grouping the DataFrame by 'Category' and calculating various aggregations

df_aggregate = txs.groupby('cust').agg({'txn_amount':['sum','mean','median','min','max']})

df_aggregate
```
